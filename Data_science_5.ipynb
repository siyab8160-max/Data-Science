```python
import numpy as np
# a=np.random.randint(1,91,25)
# b=a.reshape((5,5))
b=np.random.randint(1,91,size=(5,5))
print(b)

#row wise mean
print(np.mean(b,axis=0))
print(np.mean(b,axis=1))


# Syntax: np.where(condition, value_if_true, value_if_false) -> forms new matrix
b[b%2==0]=-1
print(b)

c=np.sort(b,axis=0)
print(c)

d=np.sort(b,axis=1)
print(d)

e=np.sum(b,axis=0)
print(e)

f=np.sum(b,axis=1)
print(f)

g=np.count_nonzero(b<0, axis=0)
print(g)

#or
count_col=np.sum(np.where(b<0,1,0),axis=0)
print(count_col)
```

```python
import numpy as np
stock = np.random.randint(1,100,size=(5,5))
print(stock)
demand= np.random.randint(1,100,size=(5,5))
print(demand)
stock=stock-demand
print(stock)

total_shortage_daywise = abs(np.sum(np.where(stock<0,stock,0),axis=1))
print(total_shortage_daywise)

stock[stock<0]=0
print(stock)

products_with_zerostock_daywise= np.sum(np.where(stock==0,1,0),axis=1)
print(products_with_zerostock_daywise)

stock=np.sort(stock,axis=1)
print(stock)
```
