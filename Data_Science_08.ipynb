JOIN

-> needs two datasets

-> can be used on one dataset as well

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2=pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})

# INNER JOIN -> fetches common values.
# pd.merge( left df, right df, common column, which join?)
df3 = pd.merge(df1,df2,on='id',how='inner')
print(df3)

# to rename the columns
df3.rename(columns={'name_x':'table1_name','name_y':'table2_name','age_x':'table1_age','age_y':'table2_age'},inplace=True)
print(df3)
```

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2=pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})

# LEFT JOIN
# pd.merge( left df, right df, common column, which join?)
df3 = pd.merge(df1,df2,on='id',how='left')
print(df3)
```

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2=pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})

# RIGHT JOIN
# pd.merge( left df, right df, common column, which join?)
df3 = pd.merge(df1,df2,on='id',how='right')
print(df3)
```

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2=pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})

# OUTER JOIN
# pd.merge( left df, right df, common column, which join?)
df3 = pd.merge(df1,df2,on='id',how='outer')
print(df3)
```

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2=pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})

# CROSS JOIN -> No 'on' parameter
# pd.merge( left df, right df, common column, which join?)
df3 = pd.merge(df1, df2, how='cross')
display(df3)
```

```python
# 183 -> LEETCODE
import pandas as pd

def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    df = pd.merge(customers,orders, left_on='id', right_on='customerId', how='left')
    df = df[df['customerId'].isna()] #null values
    df = df[['name']] #filtering only names
    df = df.rename(columns={'name':'customers'}) # renaming the name column with customers
    return df
```

```python
# 1148 -> LEETCODE
import pandas as pd

def article_views(views: pd.DataFrame) -> pd.DataFrame:
    df = views[views['author_id']==views['viewer_id']]#filtering
    df = df[['author_id']]
    sr = df['author_id'].unique()
    df = pd.DataFrame({'id': sr})
    df.sort_values('id', inplace=True)
    return df

```
