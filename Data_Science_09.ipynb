```python
class complex :
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self,other):
    a = self.x + other.x
    b = self.y + other.y
    return complex (a,b)

  def display(self):
    print(self.x, '+',self.y,'i')
a = complex(2,3)
b = complex(4,5)
c = a + b
c.display()
```
